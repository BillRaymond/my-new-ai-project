## This model model will accept X and Y coordinates
When we submit a new X value, the model will predict the probably of the Y value

```python
print("Hello World")
```

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

### üõ†Ô∏è Model preparedness
1. Create a model
2. Define the training dataset
3. Create the model


```python
# Create the model
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[1])])
model.compile(optimizer='sgd', loss='mean_squared_error')

# Define X and Y
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-2.0, 1.0, 4.0, 7.0, 10.0, 13.0], dtype=float)

# Define a model 
model.fit(xs, ys, epochs=500)
```

## üöÄ Model prediction

```python

# Predict what the value of Y will be
# is the next X value is 10
prediction = model.predict([10.0])
print("Prediction: ", prediction)
```
